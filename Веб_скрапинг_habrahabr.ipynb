```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

import pandas as pd
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError 

labels = ['Заголовок', 'Автор', 'Теги', 'Дата', 'Ссылка'] # шапка таблицы
table = pd.DataFrame(columns=labels).to_csv("data.csv", index=False) # создаем файл 

table_res = pd.read_csv('data.csv')

# массивы для хранения информации
titleList = []
authorList = []
tagsList = []
dateList = []
linkList = [] 
    
for i in range(1, 5):
  # обрабатываем исключение
  try: 
    html = urlopen(f'https://habr.com/ru/search/page{i}/?q=web%20scraping&target_type=posts&order=relevance')
  except HTTPError as e:
    print("The server returned an HTTP error")
  except URLError as e:
    print("The server could not be found!")

  
  bs = BeautifulSoup(html, "html.parser")

# парсим даннные 
  titles = bs.findAll('a', {'class': 'tm-article-snippet__title-link'})
  authors = bs.findAll('a', {'class': 'tm-user-info__username'})
  tags = bs.findAll('div', {'class': 'tm-article-snippet__hubs'})
  dates = bs.findAll('span', {'class': 'tm-article-snippet__datetime-published'})
  links = bs.findAll('a', {'class': 'tm-article-snippet__title-link'})


  for title, author, tag, date, link in zip(titles, authors, tags, dates, links):
      p = link.get('href')
      titleList.append(title.get_text())
      authorList.append(author.get_text().replace("\n","")) # удаляем переносы в строке
      tagsList.append(tag.get_text())
      dateList.append(date.get_text())
      linkList.append(f"https://habr.com{p}\n".replace("\n",""))

# сохраняем данные в таблицу
table_res['Заголовок'] = titleList
table_res['Автор'] = authorList
table_res['Теги'] = tagsList
table_res['Дата'] = dateList
table_res['Ссылка'] = linkList

table_res.to_csv('data.csv', encoding="utf-8-sig") # сохраняем в csv файл

table_res.head(50) # выводим таблицу на экран
```
